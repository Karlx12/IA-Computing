```python
import pandas as pd 
```

```python
FOLDER_DATA="./data/"
FOLDER_OUTPUT="./data/output/"
FOLDER_MODELS="./models/"
```

```python
# Lectura desde archivo de texto plano indicando el esquema de metadatos
dfpRaw = pd.read_csv(
    FOLDER_DATA+"2022-Gasto-COVID-19.csv",
    delimiter=",",
    dtype={
        "ANO_EJE": float,
        "NIVEL_GOBIERNO": str,
        "SECTOR": str,
        "PLIEGO": float,
        "MONTO_PIM": float,
        "MONTO_DEVENGADO": float,
        "MONTO_GIRADO": float
    }
)
dfpRaw
```

```python
# Conversión de variables categóricas a variables numéricas en columnas dummy
dfpRaw1 = pd.get_dummies(dfpRaw, columns=["SECTOR"])

#Verificamos
dfpRaw1
```

```python
#Definimos las columnas por normalizar
columnasFeatures = [
     "ANO_EJE",
      "SECTOR_12", "SECTOR_13", "SECTOR_22", "SECTOR_26", "SECTOR_36",
    "SECTOR_37", "SECTOR_38", "SECTOR_39", "SECTOR_40", "SECTOR_99",
      "PLIEGO",
      "MONTO_PIM",
      "MONTO_GIRADO"
]
```

```python
# Creamos un array con los nombres de las columnas normalizadas
columnasNormalizadas = []

# Iteramos y definimos los nombres de columnas
for columna in columnasFeatures:
    columnasNormalizadas.append("feature_" + columna + "_normalized")

#Verificamos
dfpRaw1[columnasFeatures]
```

```python
#Utilitario para normalizar los datos
from sklearn.preprocessing import MinMaxScaler
```

```python
#Generamos el objeto escalador con los datos
#El objeto escalador reconoce los mínimos y máximos de cada columna
#Colocará los objetos escalados en la columna "features_normalizados"
escalador = MinMaxScaler().fit(dfpRaw1[columnasFeatures])
```

```python
#Dentro del escalador se han encontrado los valores mínimos
escalador.data_min_
```

```python
#Dentro del escalador se han encontrado los valores máximos
escalador.data_max_
```

```python
dfpRaw2=dfpRaw1.copy()
#Normalizamos las columnas features y las dejamos en las columnas "feature_normalized"
dfpRaw2[columnasNormalizadas] = escalador.transform(dfpRaw2[columnasFeatures])

#Verificamos
dfpRaw2[columnasNormalizadas]
```

```python
# Almacenamos el dataframe listo para ser analizado
# Lo guardamos en un formato binario de rápido procesamiento (h5)
# El parámetro "key" hace referencia al nombre de la variable interna en el archivo binario donde se guarda la información
dfpRaw2.to_hdf(FOLDER_MODELS+"dataset_covid19_gasto.h5", key='dataset', mode='w')
```

```python
import pickle
# Almacenamos el escalador
with open(FOLDER_DATA+"escalador_covid19.pkl", "wb") as archivo:
    pickle.dump(escalador, archivo)
```
